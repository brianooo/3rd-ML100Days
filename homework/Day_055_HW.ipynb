- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
iris.feature_names
```

```python
import pandas as pd
X = pd.DataFrame(X, columns=iris.feature_names)
y = pd.DataFrame(y, columns=['class'])
X.head(), y.head()
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(X)
plt.show()
```

```python
%matplotlib inline
import numpy as np
from sklearn.cluster import KMeans

k_mean_6_bad = KMeans(n_clusters=6, n_init=1)
k_mean_3_bad = KMeans(n_clusters=3, n_init=1)
k_mean_6_goo = KMeans(n_clusters=6)
k_mean_3_goo = KMeans(n_clusters=3)

K_est = [k_mean_6_bad, k_mean_3_bad, k_mean_6_goo, k_mean_3_goo]
n_color = [6, 3, 6, 3]

for i in range(len(K_est)):
    fig = plt.figure(i+1, figsize=(6, 4))
    est = K_est[i]
    est.fit(X)
    pred = est.predict(X)
    labels = est.labels_
    sns.scatterplot('petal length (cm)', 'sepal length (cm)', hue=pred, data=X, palette=sns.color_palette(n_colors=n_color[i]))
```
